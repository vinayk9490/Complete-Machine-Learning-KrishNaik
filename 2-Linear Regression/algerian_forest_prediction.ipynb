Feature Selection Based on Correlation

```python
X_train.corr()
```

```python
#Check for Multi-Collinearity
plt.figure(figsize=(12,10))
corr = X_train.corr()
sns.heatmap(corr, annot=True)
```

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]>threshold):
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X_train, 0.80)
```

```python
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1, inplace=True)
X_train.shape, X_test.shape
```

Feature Scaling or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
X_train_scaled, X_test_scaled
```

BoxPlot to understand the effect of StandardScaler

```python
plt.subplots(figsize=(18,8))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title("X-train before Scaling")
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title("X-train After Scaling")
```

Applying various Linear Regression Models

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train_scaled,y_train)

y_pred = linreg.predict(X_test_scaled)

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
mae = mean_absolute_error(y_pred,y_test)
r2_score = r2_score(y_test, y_pred)

print(f"The Mean Absolute Error is {mae}")
print(f"The R2-Score is {r2_score}")
plt.scatter(y_test,y_pred)
```

Apply Lasso Regression to reduce overfitting 

```python
from sklearn.linear_model import Lasso
lassoreg = Lasso()
lassoreg.fit(X_train_scaled,y_train)

y_pred = lassoreg.predict(X_test_scaled)

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
mae = mean_absolute_error(y_pred,y_test)
r2_score = r2_score(y_test, y_pred)

print(f"The Mean Absolute Error is {mae}")
print(f"The R2-Score is {r2_score}")
plt.scatter(y_test,y_pred)
```

What MAE = 0.96 REALLY Means

It means:
On average, your prediction is off by 0.96 units from the actual value.
✔️ Same unit as your target variable
✔️ No percentage involved

🧠 Simple Example
Let’s say:
Actual	Predicted	Error
10	        9	      1
5	        6	      1
8	        7.2	     0.8

👉 Average error ≈ 0.96

So:
MAE = 0.96 → “I am off by ~1 unit on average”
