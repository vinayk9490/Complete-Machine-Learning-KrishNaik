```python
import numpy as np
import pandas as pd

np.random.seed(123)

n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0, n_class_1
```

```python
#Create my dataframe with imbalanced dataset

class_0 = pd.DataFrame({
    'feature1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})
```

```python
df = pd.concat([class_0, class_1], ignore_index=True)
df.tail()
```

```python
df['target'].value_counts()
```

UP-SAMPLING 

```python
df_minority = df[df['target'] == 1]
df_majority = df[df['target'] == 0]
```

```python
from sklearn.utils import resample
df_minority_upsampled = resample(df_minority, 
                                 replace=True, 
                                 n_samples=len(df_majority), 
                                 random_state=123)
```

```python
df_upsampled = pd.concat([df_majority, df_minority_upsampled], ignore_index=True)
df_upsampled['target'].value_counts()
```
